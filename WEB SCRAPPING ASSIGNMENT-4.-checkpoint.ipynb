```python
Rank=[]
State =[]
GDP=[]
GSDP_Current=[]
GSDP_Previous=[]
Share=[]
```

```python
#scraping the Rank 
r=driver.find_elements_by_xpath("//td[@class='data1']")
for i in r:
    if i.text is None :
        Rank.append("--") 
    else:
        Rank.append(i.text)
St=driver.find_elements_by_xpath("//td[@class='name']")
for i in St:
    if i.text is None :
        State.append("--") 
    else:
        State.append(i.text)
gdp=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
for i in gdp:
    if i.text is None :
        GDP.append("--") 
    else:
        GDP.append(i.text)
#scraping the Share 
shr=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]")
for i in shr:
    if i.text is None :
        Share.append("--") 
    else:
        Share.append(i.text)
#scraping the GSDP_Current 
shr=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
for i in shr:
    if i.text is None :
        GSDP_Current.append("--") 
    else:
        GSDP_Current.append(i.text)
#scraping the GSDP_Previous 
shr=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[8]")
for i in shr:
    if i.text is None :
        GSDP_Previous.append("--") 
    else:
        GSDP_Previous.append(i.text)
```

```python
#Creating Dataframe
State_GDP=pd.DataFrame([])
State_GDP['Rank']=Rank[:33]
State_GDP['State']=State[:33]
State_GDP['Share In GDP']=Share[:33]
State_GDP['GDP of State']=GDP[:33]
State_GDP['GSDP_Current']=GSDP_Current[:33]
State_GDP['GSDP_Previous']=GSDP_Previous[:33]
State_GDP        
```

6. Scrape the details of top 100 songs on billiboard.com.
Url = https:/www.billboard.com/
You have to find the following details:

A) Song name

B) Artist name

C) Last week rank

D) Peak rank

E) Weeks on board

Note: - From the home page you have to click on the charts option then hot 100-page link through code.

```python
driver = webdriver.Chrome("chromedriver.exe")
url =  "https://www.billboard.com/charts/hot-100"
driver.get(url)

Song_Name =[]
Singer=[]
rank=[]
Last_Week=[]
Weeks_on_board=[]

#scraping the Song_Name 
son=driver.find_elements_by_xpath("//li[@Class='lrv-u-width-100p']")
for i in son:
    if i.text is None :
        Song_Name.append("--") 
        Singer.append("--") 
        rank.append("--") 
        Last_Week.append("--") 
        Weeks_on_board.append("--") 
    else:
        Song_Name.append(i.text.split('\n')[0])
        Singer.append(i.text.split('\n')[1])
        rank.append(i.text.split('\n')[2])
        Last_Week.append(i.text.split('\n')[3])
        Weeks_on_board.append(i.text.split('\n')[4])


BillBOARD_HOT100=pd.DataFrame([])
BillBOARD_HOT100['Song_Name']=Song_Name
BillBOARD_HOT100['Peak_rank']=rank
BillBOARD_HOT100['Singer / Crew']=Singer
BillBOARD_HOT100['Last_Week_Rank']=Last_Week
BillBOARD_HOT100['Weeks_on_board']=Weeks_on_board

BillBOARD_HOT100
```
